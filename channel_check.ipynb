```python
path_ = './datasets/physgen/cleaned/train/'
hdf5_sample = path_ + np.random.choice(os.listdir(path_), 1)[0]
hdf5_sample
```

```python
import h5py

with h5py.File(hdf5_sample, "r") as f:
    print(f"Keys: {list(f.keys())}")
    print(f"Amount of data: {len(list(f.values()))}")

    data = list(f.values())[0][:]
    print(f"Shape: {data.shape}")

    print("\nGoing through all key-value pairs...")
    for key, value in f.items():
        print(f"Key='{key}':\n    - Value-Shape: {value.shape}\n    - Value-Type: {type(value)}")
```
